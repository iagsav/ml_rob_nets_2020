1.Визуализируйте корреляции между признаками, находящимися в наборе данных (heatmap). Сделайте выводы.

```python
correlation_heatmap = df_2017.drop(['Country', 'Happiness.Rank', 'Happiness.Score'], axis=1).corr()
sns.heatmap(correlation_heatmap, annot=True, fmt='.1f', linewidths = .5)
plt.plot()
plt.show()
```

Из данной таблицы и её визуализации можно сделать вывод, что поля: Whisker.high и Whisker.low коррелируют, т.е. взаимосвязаны друг с другом. Некоторые признаки имеют достаточно сильную взаимосвязь: Health..Life.Expectancy и Economy..GDP.per.Capita. Другие почти несвязанны, например: Dystopia.Residual и Generosity.

2.Постройте график счастья по регионам, страны на графике должны представляться отдельными точками.

```python
df_all = pd.DataFrame(df_2015.sort_values(by='Region', ascending=False))
df_all = df_all.reset_index(drop=True)
plt.ylabel('Happiness Score')
x = 0
region = df_all['Region'][0]
_x = [0]
labels = [df_all['Region'][0]]
for i in range(0, df_2015.last_valid_index()+1):
    if region is not df_all['Region'][i]:
        labels.append(df_all['Region'][i])
        region = df_all['Region'][i]
        x = x + 1
        _x.append(x)
    plt.plot(x, df_all['Happiness Score'][i], 'bo')
plt.xticks(_x, labels, rotation='vertical')
plt.show()
```

3.Оцените количество счастья по годам по регионам.

```python
df_2015 = df_2015.assign(year=pd.Series(2015, index=df_2015.index))
df_2016 = df_2016.assign(year=pd.Series(2016, index=df_2016.index))
df_all = pd.concat([df_2015, df_2016])
df_all = pd.DataFrame(df_all.sort_values(by='Country', ascending=True))
df_all = df_all.reset_index(drop=True)
df_all = pd.DataFrame(df_all.sort_values(by=['Region', 'Country'], ascending=[True, True]))
df_all = df_all.reset_index(drop=True)
mask = df_all.Country.groupby(df_all.Country).transform('value_counts') > 1
df_all = df_all[mask]
df_all[['Region', 'Happiness Score','Country', 'year']]
```
